```python
from gower.gower_dist import *
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv("SubwayMenuNutrition.csv").fillna(0)
for column in data.columns[3:-1]:
    data[column + " per calorie"] = data[column] / data["Calories"]

vectorizer = CountVectorizer(binary=True, min_df=2, stop_words="english")
X = vectorizer.fit_transform(data.Item)
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
data = pd.concat([data, df], axis=1)
data.set_index("Item", inplace=True)

data
```

```python
df, (_, res) = optimize_dbscan(data.drop("Category", axis=1), "Subway menu", data.Category,
                               factor=0.25, precomputed=True, chunksize=20)

print(res)
df
```
