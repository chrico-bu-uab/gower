```python
from pmlb import fetch_data

from gower.gower_dist import *

classification_dataset_names = ["penguins", "iris", "optdigits", "new_thyroid", "confidence", "soybean", "dermatology",
                                "mfeat_factors", "mfeat_karhunen", "segmentation", "mfeat_pixel", "ecoli", "breast",
                                "wine_recognition", "texture", "led7", "tokyo1", "mushroom", "satimage", "glass",
                                "ionosphere", "movement_libras", "house_votes_84", "analcatdata_authorship", "vowel",
                                "ring", "irish", "biomed", "breast_cancer_wisconsin"]

max_muti = 0
max_rand = 0
results = 0
for classification_dataset in classification_dataset_names:
    X, y = fetch_data(classification_dataset, return_X_y=True,
                      local_cache_dir='/Users/temp/pmlb_data')
    _, (_, res) = optimize_dbscan(pd.DataFrame(X), classification_dataset, pd.Series(y), n_iter=100,
                                  factor=0.25, precomputed=True, chunksize=20)
    print(res)
    max_muti += res.max_muti
    max_rand += res.max_rand
    results += res

max_muti /= len(classification_dataset_names)
max_rand /= len(classification_dataset_names)
results /= len(classification_dataset_names)

print("Average Mutual Information:", max_muti)
print("Average Rand Index:", max_rand)

results
```
