```python
from sklearn.feature_extraction.text import TfidfVectorizer
from gower.gower_dist import *

data = pd.read_csv('polymers_dataset.csv', index_col=0)  # .drop('smiles', axis=1)

vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 2), norm='l1')
X = vectorizer.fit_transform(data.smiles)
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
data = pd.concat([data, df], axis=1).drop('smiles', axis=1)

data = data.sample(1000, random_state=42)

data
```

```python
optimize_agglo(data.drop("label", axis=1), "polymers", data.label, precomputed=True, chunksize=20)
```
