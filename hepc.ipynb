```python
import optuna

from gower.gower_dist import *
```

```python
hepatitis = pd.read_csv('hcvdat0.csv', index_col=0)
hepatitis.fillna(hepatitis.median(), inplace=True)

def objective(trial):
    lower_q = trial.suggest_float("lower_q", 0.0, 0.5)
    c_t = trial.suggest_float("c_t", 0.0, 2.0)
    knn = trial.suggest_categorical("knn", [False, True])
    factor = trial.suggest_float("factor", 0.5, 2, log=True)
    df, niceness = optimize_dbscan(hepatitis.drop("Category", axis=1),
                                   hepatitis["Category"],
                                   factor=factor,
                                   precomputed=True,
                                   chunksize=20,
                                   lower_q=lower_q,
                                   c_t=c_t,
                                   knn=knn)
    return niceness


study = optuna.create_study(direction="maximize", sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(objective, n_trials=10)
```
